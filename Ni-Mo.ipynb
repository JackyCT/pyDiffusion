# Diffusion Simulation of Ni-Mo for 1000 hours
## Read diffusivity data from csv file

```python
%matplotlib notebook
import numpy as np
from pydiffusion import io
from pydiffusion.plot import DCplot, profileplot

profile, diffsys =  io.read_csv('Ni-Mo 1100C 1000hrs.csv')
DCplot(diffsys)
```

## Meshing
Meshing is based on similar simulated profile. The grid size $d=DC^{0.3}$. Grid number is from 300 to 400.

```python
from pydiffusion.utils import mesh, step, matanocalc

dism = mesh(profile, diffsys, [500, 600])
```

## Calculate Matano plane and set up a step profile.

```python
m = matanocalc(profile)
diffsim = step(dism, m, diffsys)
```

## Diffusion Simulation

```python
from pydiffusion.simulation import mphSim

mphSim(diffsim, diffsys, 1000*3600)
```

## Plot result

```python
profileplot(diffsim)
```
